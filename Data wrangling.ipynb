```python
df_1 = df
df_1['landing_success'] = df['landing_success'].fillna(value='False')
df_1['success'] = df['success'].fillna(value='False')
df_1.isna().mean()
```

```python
payload_type = ['Crew Dragon','Dragon 1.0','Dragon 1.1','Dragon 2.0','Satellite']
df_1[df_1['PayloadType']=='Satellite']['PayloadMass'].isna()
```

```python
for i in payload_type:
    indextp = df_1[df_1['PayloadType']==i]['PayloadMass'][df_1[df_1['PayloadType']==i]['PayloadMass'].isna()].index
    meantp = df_1[df_1['PayloadType']==i]['PayloadMass'].mean()
    df_1.loc[indextp,'PayloadMass'] = meantp
```
