```python
import folium
import pandas as pd
df = pd.read_csv('./df_final.csv')
```

```python
from folium import plugins
# [34.632093,-120.610829] 'VAFB SLC 4E'
m_VAFB = folium.Map(
    [34.632093,-120.610829],
    zoom_start= 16,control_scale=True
)

folium.Marker(
    location = [34.632093,-120.610829],
    popup='VAFB SLC 4E',
    icon=folium.Icon(color='pink')
).add_to(m_VAFB)

folium.Circle(
    radius=50,
    location = [34.631,-120.610829],
    popup='VAFB SLC 4E success',
    color = "#00CC66",
    fill = True
).add_to(m_VAFB)

folium.Circle(
    radius=50,
    location = [34.633,-120.610829],
    popup='VAFB SLC 4E False',
    color = "#FF0000",
    fill = True
).add_to(m_VAFB)

marker1 = plugins.MarkerCluster().add_to(m_VAFB)
marker2 = plugins.MarkerCluster().add_to(m_VAFB)

for j in range(len(df[df['launchpad_name']=='VAFB SLC 4E'])):
    if df[df['launchpad_name']=='VAFB SLC 4E'].iloc[j].success:
        folium.Marker([34.631,-120.610829]).add_to(marker1)
    else:
        folium.Marker([34.633,-120.610829]).add_to(marker2)
m_VAFB.save('VAFB.html')
m_VAFB
```
